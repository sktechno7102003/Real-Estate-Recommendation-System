```python
# Working on Society column

df['society'].value_counts()
```

```python
# Removing numbers, starts 

import re 
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['price'].value_counts()
```

```python
# Remobing the rows having price on request in price column

df = df[df['price'] != 'Price on Request']
```

```python
# Function to convert all price to crores

def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.head()
```

```python
# price_per_sqft column
df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

```python
df['price_per_sqft'].value_counts()
```

```python
# areaWithType column

df['areaWithType'].value_counts()
```

```python
#bedRoom column
df['bedRoom'].value_counts()
```

```python
# Removing all rows which don't have bedRoom Info bcz there other info are also not there

df = df[~df['bedRoom'].isnull()]
```

```python
# Just keep the integer value

df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).astype('int')
```

```python
df.head(5)
```

```python
# bathroom column
df[df['bathroom'].isnull()].value_counts().shape
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype('int')
```

```python
df.sample(5)
```

```python
# balcony column
df['balcony'].value_counts()
```

```python
# No balcony ko 0 and balcony word hatao

df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No', '0')
```

```python
#Extracting floor number and converting ground, basement to numbers

df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace('Ground', '0').str.replace('Basement','-1').str.replace('Lower', '0').str.extract(r'(\d+)')
```
